```python
from langchain.document_loaders import UnstructuredPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings, OpenAIEmbeddings
from langchain.vectorstores import FAISS
from langchain.chains.question_answering import load_qa_chain
from langchain import HuggingFaceHub
from langchain.llms import OpenAI

```

```python
loader = UnstructuredPDFLoader("/content/merged_files.pdf")
```

```python
document = loader.load()
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 500, chunk_overlap=0)
chunks = text_splitter.split_documents(document)
```

```python
#embedding = HuggingFaceEmbeddings()
embedding = OpenAIEmbeddings()
db = FAISS.from_documents(chunks, embedding)
```

```python
#llm=HuggingFaceHub(
#    repo_id="google/flan-t5-large",
#    model_kwargs={"temperature":0, "max_length":500}
# )

chain = load_qa_chain(OpenAI(temperature=0), chain_type="stuff")
docs2 = db.similarity_search(translated)
output_mssg = chain.run(input_documents=docs2, question=translated)
print(output_mssg)
```

```python
translated_output = GoogleTranslator(source='auto', target='ta').translate(output_mssg)
print(translated_output)
```
